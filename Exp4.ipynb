```python
#Creation of Set
thisset = {"apple", "banana", "cherry"}
print(thisset)
```

```python
#Duplicates are not allowed
thisset = {"apple", "banana", "cherry","apple"}
print(thisset)
```

```python
# The values True and 1 are considered the same value
thisset = {"apple", "banana", "cherry", True, 1, 2}
print(thisset)
```

```python
#Once a set is created , you cannot replace its items,
#but you can add new items.
thisset = {"apple", "banana", "cherry"}
thisset.add(2)

thisset.remove("banana")
print(thisset)

```

```python
# Set Operations
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

print("Union:", A | B)
print("Intersection:", A & B)
print("Difference1:", A - B)
print("Difference2:", B - A)
print("Symmetric Difference:", A ^ B)
```

```python
#frozenset
# Unlike sets , elements cannot be added or removed from a fozenset.
x = frozenset({"apple", "banana", "cherry"})
print(x)
print(type(x))
```
